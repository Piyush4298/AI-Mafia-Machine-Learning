```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Get training data set 
dfx = pd.read_csv('linearX.csv')
dfy = pd.read_csv('linearY.csv')
#changing in numpy array
dfx = dfx.values
dfy = dfy.values

x = dfx.reshape((-1,))
y = dfy.reshape((-1,))
print(x.shape,y.shape)
print(x)
```

# Data Preprocessing and Visualization

```python

x = (x-x.mean())/x.std()
y = y
plt.scatter(x,y)
plt.show
```

# Gradient Descent Algorithm
- Start with a random theta
- Repeat until converge
    - Update theta according to rule

```python
def hypothesis(x,theta):
    return theta[0]+ theta[1]*x
```

```python
def error(x,y,theta):
    error = 0
    for i in range(x.shape[0]):
        hx = hypothesis(x[i], theta)
        error += (hx-y[i])**2
    return error
```

```python
def gradient(x,y,theta):
    grad= np.zeros((2,))
    for i in range(x.shape[0]):
        hx = hypothesis(x[i], theta)
        grad[0] += (hx - y[i])
        grad[1] += (hx - y[i])*x[i]
    return grad
```

```python
def gradientDescent(x,y,learning_rate=0.001):
    theta = np.array([-2.0,0.0])
    max_iter = 100
    itr = 0
    error_list = []
    theta_list = []
    while itr<= max_iter:
        grad = gradient(x, y, theta)
        err = error(x,y,theta)
        error_list.append(err)
        theta_list.append(theta)
        theta[0] -= learning_rate*grad[0]
        theta[1] -= learning_rate*grad[1]
        
        itr += 1
    return theta, error_list, theta_list
```

```python
final_theta, error_list, theta_list = gradientDescent(x,y)
```

```python
plt.plot(error_list)
plt.show()
```

```python
print(final_theta)
```

```python
# plot line for testing data
xtest = np.linspace(-2,6,10)
print(xtest)
```

```python
plt.scatter(x,y,label="Training Data")
plt.plot(xtest, hypothesis(xtest,final_theta), color='red',label='prediction')
plt.legend()
plt.show()
```
