# KNN algorithm

```python
#Step 1: To find KNN based on euclidians distance:-
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def knn(x, y, querypoint, k=5):
    vals= []
    for i in range(x.shape[0]):
        d = distance(querypoint, x[i])
        vals.append((d, y[i]))
    #print(vals)
# Step2 : Sort the array and find k nearest points 
    vals = sorted(vals)
    vals = vals[:k]
# Majority vote
    vals = np.array(vals)
    new_values = np.unique(vals[:, 1], return_counts=True)
# index of max count:
    index = new_values[1].argmax()
#map index with data:
    pred = new_values[0][index]
    
    return pred
```

```python
knn(x, y, query)
```
